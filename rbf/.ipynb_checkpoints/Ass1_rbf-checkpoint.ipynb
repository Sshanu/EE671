```python
import sys
sys.path.insert(0, '../')
import copy
from numpy.linalg import norm
import numpy as np
from rbf_momentum import feedforward, backpropagation, loss, model
from random import randint
import pickle
import time
from matplotlib import pyplot

import time
import pickle

def load_model(name):
    f = open("save weights/rbf/"+name, 'rb')
    return pickle.load(f)
    
def save_model(name, w):
    f = open("save weights/rbf/"+name, 'wb')
    pickle.dump(w, f)
```

```python
f = open('data.txt', 'r')

data_x = []
data_y = []

for i, tt in enumerate(f):
    if(i>0):
        tmp = tt.split()
        if(i==1):
            prev = [float(tmp[0]), float(tmp[1])] 
        data_x.append([float(tmp[2]), float(tmp[3]), float(tmp[4]), prev[0], prev[1]])
        data_y.append([float(tmp[0]), float(tmp[1])])
        prev = data_y[-1]         

def normalize(t, max_, min_):
    return (t - min_) / (max_ - min_)

train_x = np.array(data_x, dtype=float)
train_y = np.array(data_y, dtype=float) + 0.4
max1, max2, max3, max4, max5 = max(train_x[:, 0]), max(train_x[:, 1]), max(train_x[:, 2]), max(train_x[:, 3]), max(train_x[:, 4])
min1, min2, min3, min4, min5 = min(train_x[:, 0]), min(train_x[:, 1]), min(train_x[:, 2]), min(train_x[:, 3]), min(train_x[:, 4])

train_x[:, 0] = normalize(train_x[:, 0], max1+0.6, min1-0.6)
train_x[:, 1] = normalize(train_x[:, 1], max2+0.6, min2-0.6)
train_x[:, 2] = normalize(train_x[:, 2], max3+0.6, min3-0.6)
train_x[:, 3] = normalize(train_x[:, 3], max4+0.02, min4-0.02)
train_x[:, 4] = normalize(train_x[:, 4], max5+0.02, min5-0.02)

print(max(train_x[:, 0]), max(train_x[:, 1]), max(train_x[:, 2]), max(train_x[:, 3]), max(train_x[:, 4]))
print(min(train_x[:, 0]), min(train_x[:, 1]), min(train_x[:, 2]), min(train_x[:, 3]), min(train_x[:, 4]))
```

```python
start = time.time()
in_dim = train_x.shape[1]
out_dim = train_y.shape[1]
layer = [100, out_dim]
learning_rate = 0.0001
learning_rate_c = 0.0001
alpha = 0.8
alpha_c = 0.8
l2_lambda = 0
batch_size = 4999
num_batch = int(train_x.shape[0]/ batch_size)
```

### Training

```python
np.random.seed(0)
init_center = copy.deepcopy(train_x)
centers, sigmas, weights, bias = model(init_center, layer)

loss_collection = []
for i in range(5000):
    for rr in range(num_batch):
        input_x = train_x[batch_size*rr:batch_size*(rr+1)]
        input_y = train_y[batch_size*rr:batch_size*(rr+1)]

        aa = feedforward(input_x[:], centers, sigmas, weights, bias)
        
        if(i==0):
            vw_old = [0 for i in range(len(weights))]
            vb_old = [0 for i in range(len(bias))]
            vs_old = [0 for i in range(len(centers))]
            vc_old = [0 for i in range(len(sigmas))]
            
        centers, sigmas, weights, bias, vc_old, vs_old, vw_old, vb_old = backpropagation(input_y, vc_old, vs_old, vw_old, vb_old, aa, centers, sigmas, weights, bias, alpha_c, alpha, learning_rate, learning_rate_c)
    if(i%1==0):
        aa = feedforward(train_x[:], centers, sigmas, weights, bias)
        yd = aa[-1]
        lo = loss(train_y, yd)
        loss_collection.append(lo)
        print(i, lo)
        
save_model('1', [centers, sigmas, weights, bias])   
print('Time:', time.time()-start)
```

```python
time_ = [i for i in range(len(loss_collection))]
fig=pyplot.figure(figsize=(10, 6), dpi= 80, facecolor='w', edgecolor='k')
pyplot.scatter(time_, loss_collection,  s=3, color='r', label="Error vs Iterations")
pyplot.ylabel("Error", fontsize=15)
pyplot.xlabel("Iterations", fontsize=15)
pyplot.ylim(0, 5.e-4)
pyplot.legend(fontsize=12, loc='upper right')
pyplot.show()
```

```python
aa = feedforward(train_x[:], centers, sigmas, weights, bias)
yd = aa[-1]
num = len(train_x)
time_ = [i for i in range(num)]
fig=pyplot.figure(figsize=(12, 6), dpi= 80, facecolor='w', edgecolor='k')
pyplot.scatter(time_, yd[:, 0]-0.4, s=3, color='r', label="Predicted $V_x$")
pyplot.scatter(time_, train_y[:, 0]-0.4, s=3, color='b', label='Desired $V_x$')
pyplot.ylabel("$V_x$", fontsize=15)
pyplot.xlabel("Time", fontsize=15)
pyplot.legend(loc='upper right')
pyplot.show()
```

```python
fig=pyplot.figure(figsize=(12, 6), dpi= 80, facecolor='w', edgecolor='k')
pyplot.scatter(time_, yd[:, 1]-0.4, s=3, color='r', label="Predicted $V_y$")
pyplot.scatter(time_, train_y[:, 1]-0.4, s=3, color='b', label='Desired $V_y$')
pyplot.ylabel("$V_y$", fontsize=15)
pyplot.xlabel("Time", fontsize=15)
pyplot.legend(loc='upper right')
pyplot.show()
```
